```python
import pandas as pd     
```

```python
game = pd.read_csv("game_diff_upd.csv")
```

```python
game.shape
```

```python
import matplotlib.pyplot as plt 
x = game["SCORE_DIFF"]
plt.figure(figsize=(10,7)) # Make it 14x7 inch
plt.style.use('seaborn-whitegrid') # nice and clean grid
plt.hist(x, bins=40, facecolor = '#2ab0ff', edgecolor='#169acf', linewidth=0.5)
plt.xlabel('Score Difference') 
plt.ylabel('Counts') 
plt.xlim([-50, 50])
plt.title("Score Difference Distribution")
plt.show()
plt.hist(game["SCORE_DIFF"])
```

```python
upset = pd.read_csv("upset_data.csv")
```

```python
upset
```

```python
counts = upset.groupby(["YEAR"], as_index= False).sum()
```

```python
fig = plt.figure(figsize = (8, 5))
 
# creating the bar plot
plt.bar(counts["YEAR"], counts["UPSET"], color ='maroon',
        width = 0.8)
 
plt.xlabel("Year")
plt.ylabel("Number of Upsets")
plt.title("Upsets in March Madness per Year")
plt.show()
```
